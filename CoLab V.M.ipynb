```python
class Item:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def updateStock(self, stock):
        self.stock = stock

    def buyFromStock(self):
        if self.stock == 0: # if no items are available
            # not raise item exception
            pass
        self.stock -= 1 # Otherwise, the item's stock will be reduced by 1.

# Make a class for the vending machine.
class VendingMachine:

    def __init__(self):
        self.amount = 0
        self.items = [] # all of the items on this list right here

    def addItem(self, item):
        self.items.append(item) # method of adding an item to a vending machine

    # method of displaying items in a vending machine
    def showItems(self):
        print('\nitems available \n***************')

        for item in self.items: # for each item available in this vending machine
            if item.stock == 0: # if this item has no stock
                self.items.remove(item) # take this item off the display
        for item in self.items:
            print(item.name + ": " "AED", item.price) # Alternatively, print this item and include its price.

        print('***************\n')

    def addCash(self, money):
        self.amount = self.amount + money # add money

    def buyItem(self, item): # if the sum you enter is less than the price
        if self.amount < item.price:
            print('You can\'t buy this item. Insert more coins.') # Obviously, you cannot purchase this item.
        else:
            self.amount -= item.price # subtract the item cost from the available cash
            item.buyFromStock() # Use this function to reduce the item inventory by 1.
            # (What if we buy several?)
            print('You got ' +item.name)
            print('Cash remaining: ' + str(self.amount))

    def containsItem(self, wanted):
        ret = False
        for item in self.items:
            if item.name == wanted:
                ret = True
                break
        return ret

    def getItem(self, wanted):
        ret = None
        for item in self.items:
            if item.name == wanted:
                ret = item
                break
        return ret

    def insertAmountForItem(self, item):
        price = item.price
        while self.amount < price:
                self.amount = self.amount + float(input('insert ' + str(price - self.amount) + ': '))

    def checkRefund(self):
        if self.amount > 0:
            print(str(self.amount) + " refunded.")
            self.amount = 0

        print('Thank you, have a nice day!\n')


def vend():

    machine = VendingMachine()
    item1 = Item('chocolate',  3,  2)
    item2 = Item('water', 1,  1)
    item3 = Item('burger',  5,  3)
    item4 = Item('bubble gum',  0.50, 1)
    item5 = Item('orange juice',  2,  3)
    item6 = Item('hotdog',  4, 5 )
    item7 = Item('salad',  4, 4)
    machine.addItem(item1)
    machine.addItem(item2)
    machine.addItem(item3)
    machine.addItem(item4)
    machine.addItem(item5)
    machine.addItem(item6)
    machine.addItem(item7)

    print('Welcome to the vending machine!\n***************')

    continueToBuy = True
    while continueToBuy == True:
        machine.showItems()
        selected = input('select item: ')
        if machine.containsItem(selected):
            item = machine.getItem(selected)

            machine.insertAmountForItem(item)
            machine.buyItem(item)

            a = input('buy something else? (yes/no): ')
            if a == 'no':
                continueToBuy = False
                machine.checkRefund()
            else:
                continue

        else:
            print('Item not available. Select another item.')
            continue

vend()
```
